```python
import matplotlib.pyplot as plt
import numpy as np
#subplot 1 
plt.suptitle("GRAPH")
plt.subplot(2,3,1)
plt.title("Title 1")
x=np.array([10,20])
y=np.array([0,175])
plt.xlabel("x label",color='blue')
plt.ylabel("y label",color='blue')
plt.plot(x,y,'o')

#subplot 2
plt.subplot(2,3,2)
x1=np.array([3,8,1,10,2])
y1=np.array([6,1,7,4,10])
plt.plot(x1,marker='*',color='orange',linewidth=2.5)
plt.plot(y1,linestyle='dotted')

#subplot 3
plt.subplot(2,3,3)
x2=np.array([2,6,3,10])
plt.plot(x2,marker='o',mfc='yellow',mec='black')

#subplot 4
plt.subplot(2,3,4)
plt.title("Title 2")
x3=np.array([0,5])
y3=np.array([0,200])
plt.plot(x3,y3)
plt.grid()


#subplot 5
plt.subplot(2,3,5)
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y,s=sizes,alpha=0.5,c=colors)


#subplot 6
plt.subplot(2,3,6)
x = np.array(["A", "B", "C"])
y = np.array([4, 2, 8])
plt.bar(x,y,color='red',width=0.1)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 20, 20])
mylabels = ["35", "20", "20", "25"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "yellow"]
plt.pie(y, labels = mylabels, startangle = 90,explode=myexplode,shadow=True,colors=mycolors)
plt.legend()
plt.show() 
```
